```python
from sklearn.datasets import fetch_openml

X_mnist, y_mnist = fetch_openml('mnist_784', return_X_y=True, as_frame=False,
                                parser='auto')
```

```python
X_train , y_train = X_mnist[:50000], y_mnist[:50000]
X_val , y_val = X_mnist[50000:60000], y_mnist[50000:60000]
X_test ,y_test = X_mnist[60000:] , y_mnist[60000:]

```

```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.metrics import accuracy_score
import pandas as pd


svc = SVC()
dt = DecisionTreeClassifier()
rf = RandomForestClassifier()
et = ExtraTreesClassifier()


classifiers = {

    "Decision Tree": dt,
    "Random Forest": rf,
    "Extra Trees": et,
    "SVC": svc,
}
preds_df = pd.DataFrame()


results = {}
for name, clf in classifiers.items():
  clf.fit(X_train, y_train)
  y_pred = clf.predict(X_val)
  preds_df[name] = y_pred
  results[name] = accuracy_score(y_val, y_pred)
  print(results[name])

results
preds_df["targer"] = y_val
preds_df.head()


```

```python
from sklearn.ensemble import VotingClassifier


voting_clf = VotingClassifier(
    estimators=[('dt', dt), ('rf', rf), ('et', et)],
    voting='hard'
)
voting_clf.fit(X_train, y_train)
y_pred = voting_clf.predict(X_val)
accuracy_score(y_val, y_pred)
```

```python
y_pred = voting_clf.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
voting_clf.voting = "soft"
voting_clf.fit(X_train, y_train)
y_pred = voting_clf.predict(X_val)
accuracy_score(y_val, y_pred)
```

```python
from sklearn.neural_network import MLPClassifier

blender = MLPClassifier(learning_rate_init=0.01,max_iter=1000)
blender.fit(preds_df.drop('targer',axis=1),preds_df["targer"])
```

```python
preds_df2 = pd.DataFrame()
results = {}
for name, clf in classifiers.items():

  y_pred = clf.predict(X_test)
  preds_df2[name] = y_pred
  results[name] = accuracy_score(y_test, y_pred)
  print(results[name])
```

```python
y_final_predict = blender.predict(preds_df2)
accuracy_score(y_final_predict,y_test)
```
